```python
# =========================
# Imports
# =========================
import wfdb
import numpy as np
from scipy.signal import butter, lfilter
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import (
    Conv1D, MaxPooling1D, Dropout, Dense, BatchNormalization,
    Input, Bidirectional, LSTM, SpatialDropout1D, GlobalAveragePooling1D
)

from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import matplotlib.pyplot as plt
from collections import Counter

# =========================
# AAMI 5-Class Mapping
# =========================
def map_to_aami(label):
    if label in ['N', 'L', 'R', 'e', 'j']:
        return 'N'
    elif label in ['A', 'a', 'J', 'S']:
        return 'S'
    elif label in ['V', 'E']:
        return 'V'
    elif label in ['F']:
        return 'F'
    elif label in ['/', 'f', 'Q']:
        return 'Q'
    else:
        return None

# =========================
# ECG Preprocessing
# =========================
def load_and_preprocess_single_record(record_number):
    record = wfdb.rdrecord(record_number, pn_dir='mitdb/')
    annotation = wfdb.rdann(record_number, 'atr', pn_dir='mitdb/')

    signal = record.p_signal[:, 0]
    r_peaks = annotation.sample

    # Bandpass Filter
    low_cutoff, high_cutoff = 0.5, 40
    nyquist_freq = 0.5 * record.fs
    b, a = butter(4, [low_cutoff / nyquist_freq, high_cutoff / nyquist_freq], btype='band')
    filtered_signal = lfilter(b, a, signal)

    # Normalize
    scaler = MinMaxScaler()
    normalized_signal = scaler.fit_transform(filtered_signal.reshape(-1, 1)).flatten()

    # Segmentation
    segments, labels = [], []
    pre_samples = int(0.2 * record.fs)
    post_samples = int(0.3 * record.fs)

    for i in range(len(r_peaks)):
        start = r_peaks[i] - pre_samples
        end = r_peaks[i] + post_samples
        if start >= 0 and end < len(normalized_signal):
            aami_label = map_to_aami(annotation.symbol[i])
            if aami_label is not None:
                segments.append(normalized_signal[start:end])
                labels.append(aami_label)

    return np.array(segments), np.array(labels)

# =========================
# Load Data
# =========================
record_numbers = [
    '100','101','102','103','104','105','106','107','108','109','111','112',
    '113','114','115','116','117','118','119','121','122','123','124','200',
    '201','202','203','205','207','208','209','210','212','213','214','215',
    '217','219','220','221','222','223','228','230','231','232','233','234'
]

all_X, all_y = [], []
for rec in record_numbers:
    X, y = load_and_preprocess_single_record(rec)
    all_X.append(X)
    all_y.append(y)

full_X = np.concatenate(all_X)
full_y = np.concatenate(all_y)

print("\nTotal Beats:", len(full_y))
print("AAMI Class Distribution:", Counter(full_y))

# =========================
# Label Encoding
# =========================
labels = np.unique(full_y)
label_to_int = {label: i for i, label in enumerate(labels)}
y_int = np.array([label_to_int[lbl] for lbl in full_y])

# =========================
# Train-Test Split
# =========================
X_train, X_test, y_train_int, y_test_int = train_test_split(
    full_X, y_int, test_size=0.2, stratify=y_int, random_state=42
)

X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)

num_classes = len(labels)
y_train_cat = to_categorical(y_train_int, num_classes)
y_test_cat = to_categorical(y_test_int, num_classes)

# =========================
# Class Weighting
# =========================
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_train_int),
    y=y_train_int
)
class_weights = {i: w for i, w in enumerate(class_weights)}
print("\nClass Weights:", class_weights)

# =========================
# CNN + BiLSTM Model
# =========================
input_layer = Input(shape=(X_train.shape[1], 1))

# CNN Feature Extraction
x = Conv1D(32, 7, padding="same", activation="relu")(input_layer)
x = BatchNormalization()(x)
x = SpatialDropout1D(0.1)(x)
x = MaxPooling1D(2)(x)

x = Conv1D(64, 5, padding="same", activation="relu")(x)
x = BatchNormalization()(x)
x = SpatialDropout1D(0.1)(x)
x = MaxPooling1D(2)(x)

# BiLSTM Sequence Learning
x = Bidirectional(LSTM(64, return_sequences=True))(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

x = Bidirectional(LSTM(64))(x)
x = BatchNormalization()(x)
x = Dropout(0.3)(x)

# FC Layers
x = Dense(128, activation="relu")(x)
x = Dropout(0.3)(x)

output = Dense(num_classes, activation="softmax")(x)

model = Model(inputs=input_layer, outputs=output)
model.compile(optimizer=Adam(0.001), loss="categorical_crossentropy", metrics=["accuracy"])
model.summary()

# =========================
# Training Setup
# =========================
save_path = "best_cnn_bilstm_using5classes_aami.keras"

checkpoint = ModelCheckpoint(save_path, monitor='val_loss', save_best_only=True)
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', patience=2, factor=0.5)

history = model.fit(
    X_train, y_train_cat,
    validation_split=0.15,
    epochs=50,
    batch_size=64,
    class_weight=class_weights,
    callbacks=[checkpoint, early_stop, reduce_lr],
    verbose=1
)

# =========================
# Evaluation
# =========================
model = load_model(save_path)
test_loss, test_acc = model.evaluate(X_test, y_test_cat)
print(f"\n✅ Test Accuracy: {test_acc*100:.2f}%")

y_pred = np.argmax(model.predict(X_test), axis=1)
int_to_label = {v: k for k, v in label_to_int.items()}
target_names = [int_to_label[i] for i in range(num_classes)]

print("\n✅ Classification Report:")
print(classification_report(y_test_int, y_pred, target_names=target_names))

# =========================
# Confusion Matrix
# =========================
cm = confusion_matrix(y_test_int, y_pred)
plt.figure(figsize=(8,6))
plt.imshow(cm, cmap="Blues")
plt.title("AAMI 5-Class Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.colorbar()
plt.show()

# =========================
# Training Curves
# =========================
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Accuracy")
plt.legend(["Train", "Validation"])

plt.subplot(1,2,2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Loss")
plt.legend(["Train", "Validation"])

plt.tight_layout()
plt.show()
```
