```python
# =========================
# Imports
# =========================
import wfdb
import numpy as np
from scipy.signal import butter, lfilter
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from collections import Counter
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import (
    Conv1D, MaxPooling1D, Dropout, GRU, Dense, BatchNormalization,
    Bidirectional
)
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import matplotlib.pyplot as plt
import os


# =========================
# ECG Preprocessing Function
# =========================
def load_and_preprocess_single_record(record_number, visualize=False):
    try:
        record = wfdb.rdrecord(record_number, pn_dir='mitdb/')
        annotation = wfdb.rdann(record_number, 'atr', pn_dir='mitdb/')
    except Exception as e:
        print(f"Error loading record {record_number}: {e}")
        return None, None

    signal = record.p_signal[:, 0]
    r_peaks = annotation.sample

    # Filtering
    low_cutoff, high_cutoff = 0.5, 40
    nyquist_freq = 0.5 * record.fs
    b, a = butter(4, [low_cutoff / nyquist_freq, high_cutoff / nyquist_freq], btype='band')
    filtered_signal = lfilter(b, a, signal)

    # Normalizing
    scaler = MinMaxScaler()
    normalized_signal = scaler.fit_transform(filtered_signal.reshape(-1, 1)).flatten()

    # Segmentation
    segments, labels = [], []
    pre_samples = int(0.2 * record.fs)
    post_samples = int(0.3 * record.fs)

    for i in range(len(r_peaks)):
        start = r_peaks[i] - pre_samples
        end = r_peaks[i] + post_samples
        if start >= 0 and end < len(normalized_signal):
            segments.append(normalized_signal[start:end])
            labels.append(annotation.symbol[i])

    return np.array(segments), np.array(labels)


# =========================
# Load Data
# =========================
record_numbers = [
    '100','101','102','103','104','105','106','107','108','109','111','112',
    '113','114','115','116','117','118','119','121','122','123','124','200',
    '201','202','203','205','207','208','209','210','212','213','214','215',
    '217','219','220','221','222','223','228','230','231','232','233','234'
]

all_X, all_y = [], []
print("\n--- Loading & Preprocessing Records ---")
for rec in record_numbers:
    X, y = load_and_preprocess_single_record(rec)
    if X is not None:
        all_X.append(X)
        all_y.append(y)

full_X = np.concatenate(all_X)
full_y = np.concatenate(all_y)

print(f"Total beats: {len(full_y)}")
print("Class distribution:", Counter(full_y))


# =========================
# Train-Test Split
# =========================
X_train, X_test, y_train, y_test = train_test_split(
    full_X, full_y, test_size=0.2, random_state=42, stratify=full_y
)


# =========================
# Remove Rare Classes (<6 samples)
# =========================
train_counts = Counter(y_train)
rare_classes = [cls for cls, cnt in train_counts.items() if cnt < 6]

train_mask = [lbl not in rare_classes for lbl in y_train]
test_mask = [lbl not in rare_classes for lbl in y_test]

X_train = X_train[train_mask]
y_train = y_train[train_mask]
X_test = X_test[test_mask]
y_test = y_test[test_mask]


# =========================
# Label Encoding + SMOTE
# =========================
label_to_int = {label: i for i, label in enumerate(np.unique(y_train))}
y_train_int = np.array([label_to_int[lbl] for lbl in y_train])
y_test_int = np.array([label_to_int[lbl] for lbl in y_test])

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train, y_train_int)


# Reshape
X_train_res = np.expand_dims(X_train_res, -1)
X_test = np.expand_dims(X_test, -1)

num_classes = len(np.unique(y_train_res))
y_train_cat = to_categorical(y_train_res, num_classes)
y_test_cat = to_categorical(y_test_int, num_classes)


# =========================
# CNN-GRU Model
# =========================
model = Sequential([
    Conv1D(32, 7, activation='relu', padding='same', input_shape=(X_train_res.shape[1], 1)),
    BatchNormalization(),
    MaxPooling1D(2),

    Conv1D(64, 5, activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling1D(2),
    Dropout(0.25),

    Bidirectional(GRU(64, dropout=0.2)),
    Dropout(0.4),

    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

model.compile(
    optimizer=Adam(0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()


# =========================
# Training Setup
# =========================
save_path = "/content/drive/MyDrive/best_cnn_gru_aami5.keras" \
    if os.path.exists("/content/drive/MyDrive") \
    else "best_cnn_gru_aami5.keras"

checkpoint = ModelCheckpoint(
    filepath=save_path,
    monitor='val_loss',
    save_best_only=True,
    mode='min'
)

early_stop = EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', patience=2, factor=0.5, verbose=1)


# =========================
# Train (✅ 12 Epochs)
# =========================
history = model.fit(
    X_train_res, y_train_cat,
    validation_split=0.15,
    epochs=12,
    batch_size=64,
    callbacks=[checkpoint, early_stop, reduce_lr],
    verbose=1
)

print(f"\n✅ Best model saved at: {save_path}")


# =========================
# ✅ Load Best Model & Evaluate
# =========================
model = load_model(save_path)
test_loss, test_acc = model.evaluate(X_test, y_test_cat)
print(f"\n✅ Test Accuracy: {test_acc*100:.2f}%")


# =========================
# ✅ Classification Report + Confusion Matrix
# =========================
y_pred = np.argmax(model.predict(X_test), axis=1)

int_to_label = {v: k for k, v in label_to_int.items()}
target_names = [int_to_label[i] for i in range(num_classes)]

print("\nClassification Report:")
print(classification_report(y_test_int, y_pred, target_names=target_names))

cm = confusion_matrix(y_test_int, y_pred)
plt.figure(figsize=(8,6))
plt.imshow(cm, cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.colorbar()
plt.show()


# =========================
# ✅ Training Curves
# =========================
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Accuracy Over Epochs")
plt.legend(["Train", "Validation"])

plt.subplot(1,2,2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title("Loss Over Epochs")
plt.legend(["Train", "Validation"])

plt.tight_layout()
plt.show()
```

```python
# Save the model (architecture + weights + optimizer state)
model.save("cardiac_arrhythmia_cnn_gru_model.h5")

print("✅ Model saved successfully as cardiac_arrhythmia_cnn_gru_model.h5")
```

```python
# =========================
# Install dependencies
# =========================
!pip install wfdb

# =========================
# Imports
# =========================
import wfdb
import numpy as np
from scipy.signal import butter, lfilter
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt
from collections import Counter

import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# =========================
# Custom Focal Loss
# =========================
def focal_loss(gamma=2., alpha=0.25):
    def loss(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        y_pred = tf.clip_by_value(y_pred, 1e-7, 1.0)

        ce = -y_true * tf.math.log(y_pred)
        weight = alpha * tf.math.pow((1 - y_pred), gamma)
        fl = weight * ce
        return tf.reduce_sum(fl, axis=1)
    return loss

# =========================
# AAMI 5-Class Mapping
# =========================
def map_to_aami(label):
    if label in ['N','L','R','e','j']: return 'N'
    elif label in ['A','a','J','S']: return 'S'
    elif label in ['V','E']: return 'V'
    elif label == 'F': return 'F'
    elif label in ['/', 'f', 'Q']: return 'Q'
    else: return None

# =========================
# ECG Preprocessing + RR Features
# =========================
def load_and_preprocess_single_record(record_number):
    record = wfdb.rdrecord(record_number, pn_dir='mitdb/')
    annotation = wfdb.rdann(record_number, 'atr', pn_dir='mitdb/')

    signal = record.p_signal[:,0]
    fs = record.fs
    r_peaks = annotation.sample

    # Bandpass filter
    low, high = 0.5, 40
    nyq = 0.5 * fs
    b, a = butter(4, [low/nyq, high/nyq], btype='band')
    filtered = lfilter(b, a, signal)

    # Normalize
    scaler = MinMaxScaler()
    norm_signal = scaler.fit_transform(filtered.reshape(-1,1)).flatten()

    segments, labels, rr_features = [], [], []

    pre = int(0.2 * fs)
    post = int(0.3 * fs)

    for i in range(1, len(r_peaks)-1):
        start = r_peaks[i] - pre
        end   = r_peaks[i] + post

        if start < 0 or end >= len(norm_signal):
            continue

        aami = map_to_aami(annotation.symbol[i])
        if aami is None:
            continue

        rr_prev = (r_peaks[i] - r_peaks[i-1]) / fs
        rr_next = (r_peaks[i+1] - r_peaks[i]) / fs

        rr_features.append([rr_prev, rr_next])
        segments.append(norm_signal[start:end])
        labels.append(aami)

    return np.array(segments), np.array(rr_features), np.array(labels)

# =========================
# Load ALL MIT-BIH Records
# =========================
record_numbers = [
 '100','101','102','103','104','105','106','107','108','109','111','112',
 '113','114','115','116','117','118','119','121','122','123','124','200',
 '201','202','203','205','207','208','209','210','212','213','214','215',
 '217','219','220','221','222','223','228','230','231','232','233','234'
]

all_X, all_RR, all_y = [], [], []

for rec in record_numbers:
    X, RR, y = load_and_preprocess_single_record(rec)
    all_X.append(X)
    all_RR.append(RR)
    all_y.append(y)

X = np.concatenate(all_X)
RR = np.concatenate(all_RR)
y = np.concatenate(all_y)

print("Total beats:", len(y))
print("Class distribution:", Counter(y))

# =========================
# Encode Labels
# =========================
labels = np.unique(y)
label_to_int = {lbl:i for i,lbl in enumerate(labels)}
int_to_label = {i:lbl for lbl,i in label_to_int.items()}

y_int = np.array([label_to_int[l] for l in y])

# =========================
# Train/Test Split
# =========================
X_train, X_test, RR_train, RR_test, y_train, y_test = train_test_split(
    X, RR, y_int, test_size=0.2, stratify=y_int, random_state=42
)

X_train = np.expand_dims(X_train, -1)
X_test  = np.expand_dims(X_test, -1)

num_classes = len(labels)
y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes)
y_test_cat  = tf.keras.utils.to_categorical(y_test, num_classes)

# =========================
# Build CNN-BiGRU Model
# =========================
def build_cnn_bigru(input_len, num_classes):
    ecg_in = Input(shape=(input_len, 1))
    rr_in = Input(shape=(2,))

    # ---- CNN FEATURE EXTRACTOR ----
    x = Conv1D(32, kernel_size=7, padding="same", activation="relu")(ecg_in)
    x = MaxPooling1D(2)(x)

    x = Conv1D(64, kernel_size=5, padding="same", activation="relu")(x)
    x = MaxPooling1D(2)(x)

    x = Conv1D(128, kernel_size=3, padding="same", activation="relu")(x)

    # ---- BiGRU BLOCK ----
    x = Bidirectional(GRU(64, return_sequences=True))(x)
    x = Bidirectional(GRU(32))(x)

    # ---- RR INTERVAL BRANCH ----
    rr_branch = Dense(16, activation="relu")(rr_in)

    # ---- MERGE BOTH ----
    merged = Concatenate()([x, rr_branch])
    merged = Dense(128, activation="relu")(merged)
    merged = Dropout(0.3)(merged)

    output = Dense(num_classes, activation="softmax")(merged)

    model = Model([ecg_in, rr_in], output)
    return model


model = build_cnn_bigru(X_train.shape[1], num_classes)
model.compile(optimizer=Adam(0.001), loss=focal_loss(), metrics=["accuracy"])
model.summary()

# =========================
# Training
# =========================
callbacks = [
    EarlyStopping(patience=5, restore_best_weights=True),
    ReduceLROnPlateau(patience=2, factor=0.5),
    ModelCheckpoint("best_rr_cnn_bigru.keras", save_best_only=True)
]

history = model.fit(
    [X_train, RR_train], y_train_cat,
    validation_split=0.15,
    epochs=40,
    batch_size=64,
    callbacks=callbacks,
    verbose=1
)

# =========================
# Evaluation
# =========================
model = tf.keras.models.load_model("best_rr_cnn_bigru.keras", compile=False)
model.compile(optimizer=Adam(0.001), loss=focal_loss(), metrics=["accuracy"])

loss, acc = model.evaluate([X_test, RR_test], y_test_cat)
print("\nTest Accuracy:", acc)

y_pred = np.argmax(model.predict([X_test, RR_test]), axis=1)

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=labels))

# =========================
# Confusion Matrix
# =========================
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7,5))
plt.imshow(cm, cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.colorbar()
plt.show()

# =========================
# Training Curves
# =========================
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.title("Accuracy")

plt.subplot(1,2,2)
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.title("Loss")
plt.show()
```
